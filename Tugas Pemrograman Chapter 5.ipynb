```python
import pandas as pd #for data handling
from scipy.stats.mstats import gmean, hmean #for calculating the geometric and harmonic means
import matplotlib.pyplot as plt #for plotting

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

a_mean = my_dataset.Rb.mean() #calculate the arithmetic mean using the built-in pandas function
g_mean = gmean(my_dataset['Rb']) #calculate the geometric mean using the gmean function from scipy.stats.mstats
h_mean = hmean(my_dataset['Rb']) #calculate the harmonic mean using the hmean function from scipy.stats.mstats

print ('-------') #just for better formatting of the output
print ('arithmetic mean') #print the arithmetic mean with one decimal place and the unit [ppm]
print ("{0:.1f} [ppm]".format(a_mean)) #print the arithmetic mean with one decimal place and the unit [ppm]
print ('-------') #just for better formatting of the output

print ('geometric mean') #print the geometric mean with one decimal place and the unit [ppm]
print ("{0:.1f} [ppm]".format(g_mean)) #print the geometric mean with one decimal place and the unit [ppm]
print ('-------') #just for better formatting of the output

print ('harmonic mean') #print the harmonic mean with one decimal place and the unit [ppm]
print ("{0:.1f} [ppm]".format(h_mean)) #print the harmonic mean with one decimal place and the unit [ppm]
print ('-------') #just for better formatting of the output

fig, ax = plt.subplots() #create a figure and an axes object for plotting
ax.hist(my_dataset.Rb, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) #plot a histogram of the Rb measurements with automatic binning, density normalization, black edges, a label for the legend, and some transparency
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2) #add a vertical line for the arithmetic mean with purple color, a label for the legend, and a linewidth of 2
ax.axvline(g_mean, color='orange',  label='Geometric mean', linewidth=2) #add a vertical line for the geometric mean with orange color, a label for the legend, and a linewidth of 2
ax.axvline(h_mean, color='green',  label='Harmonic mean', linewidth=2) #add a vertical line for the harmonic mean with green color, a label for the legend, and a linewidth of 2
ax.set_xlabel('Rb [ppm]') #set the label for the x-axis
ax.set_ylabel('Probability density') #set the label for the y-axis
ax.legend() #add a legend to the plot

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd #for data handling
import matplotlib.pyplot as plt #for plotting

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #read the data from the excel file

variance =  my_dataset['La'].var() #calculate the variance of the La measurements using the built-in pandas function, which by default calculates the sample variance (dividing by n-1)
stddev =  my_dataset['La'].std() #calculate the standard deviation of the La measurements using the built-in pandas function, which by default calculates the sample standard deviation (dividing by n-1)

print ('-------') #just for better formatting of the output
print ('Variance') #print the label for the variance
print("{0:.0f} [square ppm]".format(variance)) #print the variance with no decimal places and the unit [square ppm]
print ('-------') #just for better formatting of the output
print ('Standard Deviation') #print the label for the standard deviation
print("{0:.0f} [ppm]".format(stddev)) #print the standard deviation with no decimal places and the unit [ppm]
print ('-------') #just for better formatting of the output

fig, ax = plt.subplots() #create a figure and an axes object for plotting
ax.hist(my_dataset.La, bins= 20, density = True, edgecolor='k', label='Measurements Hist') #plot a histogram of the La measurements with 20 bins, density normalization, black edges, a label for the legend, and some transparency
ax.axvline(my_dataset['La'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2) #add a vertical line for the mean minus one standard deviation with purple color, a label for the legend using LaTeX formatting for the sigma symbol, and a linewidth of 2
ax.axvline(my_dataset['La'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2) #add a vertical line for the mean plus one standard deviation with green color, a label for the legend using LaTeX formatting for the sigma symbol, and a linewidth of 2
ax.axvspan(my_dataset['La'].mean()-stddev, my_dataset['La'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') #add a vertical span for the mean plus/minus one standard deviation with orange color, some transparency, and a label for the legend using LaTeX formatting for the sigma symbol
ax.set_xlabel('La [ppm]') #set the label for the x-axis
ax.set_ylabel('Probability density') #set the label for the y-axis
ax.legend() #add a legend to the plot

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd #for data handling
import numpy as np #for calculating the percentiles
import matplotlib.pyplot as plt #for plotting

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #read the data from the excel file

q1 = np.percentile(my_dataset.Nd, 25, method = 'midpoint') #calculate the first quartile (Q1) of the Nd measurements using the numpy function percentile with the method 'midpoint' for interpolation
q3 = np.percentile(my_dataset.Nd, 75, method = 'midpoint') #calculate the third quartile (Q3) of the Nd measurements using the numpy function percentile with the method 'midpoint' for interpolation
  
iqr = q3 - q1 # Interquaritle range (IQR) #calculate the interquartile range (IQR) as the difference between the third quartile (Q3) and the first quartile (Q1)

print ('-------') #just for better formatting of the output
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) #print the interquartile range (IQR) with no decimal places and the unit [ppm]
print ('-------') #just for better formatting of the output

fig, ax = plt.subplots() #create a figure and an axes object for plotting
ax.hist(my_dataset.Nd, bins='auto', density=True, edgecolor='k', label='Measurements Hist') #plot a histogram of the Nd measurements with automatic binning, density normalization, black edges, and a label for the legend
ax.axvline(q1, color='purple', label='Q1', linewidth=2) #add a vertical line for the first quartile (Q1) with purple color, a label for the legend, and a linewidth of 2
ax.axvline(q3, color='green', label='Q3', linewidth=2) #add a vertical line for the third quartile (Q3) with green color, a label for the legend, and a linewidth of 2
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') #add a vertical span for the interquartile range (IQR) with orange color, some transparency, and a label for the legend
ax.set_xlabel('Nd [ppm]') #set the label for the x-axis
ax.set_ylabel('Probability density') #set the label for the y-axis
ax.legend() #add a legend to the plot

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import numpy as np #for calculating the histogram and the bin edges
from scipy.stats import skew #for calculating the skewness of the distribution

a_mean = my_dataset.Ce.mean() #calculate the arithmetic mean using the built-in pandas function
median = my_dataset.Ce.median() #calculate the median using the built-in pandas function
hist, bin_edges = np.histogram(my_dataset['Ce'], bins=20, density=True) #calculate the histogram of the Ce measurements with 20 bins and density normalization, and also get the bin edges
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 #calculate the modal value as the midpoint of the bin with the highest frequency
standard_deviation = my_dataset['Ce'].std() #calculate the standard deviation of the Ce measurements using the built-in pandas function, which by default calculates the sample standard deviation (dividing by n-1)

a1 = (a_mean - modal_value) / standard_deviation if standard_deviation != 0 else 0 #calculate Pearson's first coefficient of skewness
a2 = 3 * (a_mean - median) / standard_deviation if standard_deviation != 0 else 0 #calculate Pearson's second coefficient of skewness
g1 = skew(my_dataset['Ce']) #calculate Fisher-Pearson's coefficient of skewness

print ('-------') #just for better formatting of the output
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1)) #print Pearson's first coefficient of skewness with two decimal places
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2)) #print Pearson's second coefficient of skewness with two decimal places
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1)) #print Fisher-Pearson's coefficient of skewness with two decimal places
print ('-------') #just for better formatting of the output

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```

```python
import pandas as pd #for data handling
import matplotlib.pyplot as plt #for plotting

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #read the data from the excel file

fig, ax = plt.subplots() #create a figure and an axes object for plotting
my_flierprops = dict(markerfacecolor="#12bbef", markeredgecolor="#61ec72", marker='o') #customize the appearance of the outliers (fliers) in the boxplot with a light yellow face color, dark blue edge color, and circular markers
my_medianprops = dict(color="#781717", linewidth = 2) #customize the appearance of the median line in the boxplot with a light red color and a linewidth of 2
my_boxprops = dict(facecolor="#88c9ec", edgecolor='#24305e') #customize the appearance of the box in the boxplot with a light blue face color and dark blue edge color
ax.boxplot(my_dataset.Nd, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops) #plot a boxplot of the Nd measurements with notches, customized appearance for the box, median line, and outliers (fliers), and patch_artist=True to fill the box with color
ax.set_ylabel('Nd [ppm]') #set the label for the y-axis
ax.set_xticks([1]) #set the x-ticks to be at position 1 (since we have only one boxplot)
ax.set_xticklabels(['all Epochs']) #set the x-tick label to 'all Epochs'
plt.show() #display the plot
```
